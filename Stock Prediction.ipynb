# TESLA STOCK PREDICTION USING MACHINE LEARNING

## Importing libraries

```python
import pandas as pd
import sklearn
import math
import numpy as np
from sklearn import preprocessing, svm
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
import matplotlib.pyplot as plt
from matplotlib import style
import datetime
style.use('ggplot')
```

## Importing data from csv file

```python
data=pd.read_csv("Tesla.csv",index_col='Date',parse_dates=True)
```

```python
data.head()
```

## Defining Label to be foreasted

```python
forecast_var = 'Adj Close'
forcast_length = int(math.ceil(len(data)*0.05))
data1 = data.copy()
```

## Creating Label column to be forecasted

```python
data["Label"] = data[forecast_var].shift(-forcast_length)
data.head()
```

## Seperating data

```python
## Seperating dependent and independent variable
X = np.array(data.drop(['Label'],axis=1))
X1 = X[-forcast_length:]
X = X[:-forcast_length]
X_full = np.array(data1)
X_full
```

```python
## Creating dependent variable
data.dropna(inplace=True)
y = np.array(data['Label'])
```

## Train Test Split

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(X,y,test_size=0.2,random_state = 0)
```

## Modeling

```python
## Using sklearn linear regression model
model = LinearRegression(n_jobs=-1)
```

```python
#fitting data in the model
model.fit(X,y)
```

## Predicting labels for test data

```python
from sklearn.metrics import r2_score
```

```python
y_test_pred = model.predict(x_test)
```

```python
score = r2_score(y_test,y_test_pred)
```

```python
print("Accuracy Score for Test Data: ",score)
```
